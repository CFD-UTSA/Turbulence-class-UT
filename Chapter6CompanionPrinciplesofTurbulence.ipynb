# 6.1 Deterministic vs. Stochastic Processes


A **deterministic process** is one in which the outcome at any time \( t \) is completely determined by the initial conditions and governing equations.  
No uncertainty is involved — given identical inputs, the system always evolves in the same way. Examples include:

- Newton’s laws of motion for an idealized particle,
- the laminar solution of the Navier–Stokes equations,
- or the exponential decay of a radioactive isotope’s *expected* population.

In contrast, a **stochastic process** (also called a *random process*) is one whose evolution in time (or space) is not uniquely determined by the initial state.  
Instead, each realization or sample path of the process represents one possible outcome drawn from an ensemble of possibilities.  
The state variable X(t) is treated as a random variable that depends on time  t .

$$\text{Mathematically, a stochastic process is a family of random variables X(t) : t in T,  
defined on a probability space:}  (\Omega, \mathcal{F}, P), \text{where each outcome}  (\omega \in \Omega ) \text{produces a distinct trajectory}  (X(t, \omega))$$

### Random Variable

The concept of **random variable** is central to stochastic modeling.  
It represents the natural variability inherent in physical or measurement processes — the fluctuations that cannot be predicted exactly but whose statistical properties can be described.

Formally, if a quantity  X  does not take a single deterministic value but rather assumes different values with certain probabilities, then  X  is said to a random variable.  
The behavior of X is characterized not by a single function but by its probability distribution, expected value, and higher-order moments.

### Relevance to Turbulence

Turbulence is a quintessential example of a stochastic process in physics:  
although the Navier–Stokes equations are deterministic, the extreme sensitivity to initial and boundary conditions makes the instantaneous velocity field unpredictable in detail.  
Therefore, we represent turbulent quantities (velocity, pressure, scalar concentration) statistically —  
as random processes whose *ensemble averages* and *correlation functions* describe the reproducible features of the flow.

---

> **Summary**
>
> - *Deterministic process*: governed by exact laws, reproducible trajectories.  
> - *Stochastic process*: governed by probability laws, each realization differs.  
> - *Random variation*: inherent fluctuation about a mean state, quantifiable by statistical descriptors such as mean, variance, and PDF.

## What the Code Does

1. **Generates time series** for several representative processes:

   - **White Noise** – a sequence of independent, zero-mean Gaussian samples with variance \( \sigma^2 \).  
     Represents completely uncorrelated fluctuations — the simplest model of turbulence noise.

   - **Ornstein–Uhlenbeck (OU) Process** – a mean-reverting Gaussian process defined by  

     $$
     x_{k+1} = x_k + \big(-\theta (x_k - \mu)\big)\,\Delta t + \sigma\,\sqrt{\Delta t}\,N(0,1),
     $$

     where \( \theta \) controls the relaxation rate.  
     This models *colored noise* with a finite correlation time, often used in Langevin models of turbulence.

   - **Gaussian Process (Integrated White Noise)** – the cumulative integral of white noise, producing a random signal with variance that grows over time.

   - **Poisson Process** – a discrete counting process with random jumps occurring at a mean rate \( \lambda \).  
     Used to model arrival events such as eddies, bursts, or particle collisions.

   - **Finite-State Markov Chain** – a simple process where the system transitions between discrete states according to a probability matrix \( P \).  
     Demonstrates the *memory property* and stationary distributions of Markov processes.

   - **Random Walk / Brownian Motion** – the cumulative sum of random steps, representing diffusive motion and serving as a limiting case of stochastic particle trajectories.

2. **Visualizes each process** in a separate plot:

   - Time series of the OU process \( x(t) \)  
   - Realization of white noise  
   - Integrated Gaussian process  
   - Poisson counting process \( N(t) \)  
   - Sequence of Markov chain states  
   - Random walk or Brownian trajectory

````python
# Example NB 6.1 Simulate canonical stochastic processes with optional ipywidgets controls
import numpy as np
import matplotlib.pyplot as plt

# Try to import ipywidgets; fall back to static parameters if unavailable
try:
    import ipywidgets as W
    HAS_WIDGETS = True
except Exception:
    HAS_WIDGETS = False

rng = np.random.default_rng(42)

def simulate_processes(T=10.0, dt=0.01, theta=1.0, sigma=1.0, mu=0.0, lam=5.0, Nstates=3):
    n = int(T/dt)
    t = np.linspace(0, T, n, endpoint=False)

    # White noise
    white = sigma * rng.standard_normal(n)

    # OU: x_{k+1} = x_k + (-theta*(x_k-mu))*dt + sigma*sqrt(dt)*N(0,1)
    x = np.zeros(n)
    x[0] = mu
    for k in range(n-1):
        x[k+1] = x[k] + (-theta*(x[k]-mu))*dt + sigma*np.sqrt(dt)*rng.standard_normal()

    # Gaussian process (integrated white noise as simple example)
    gauss_proc = np.cumsum(white) * np.sqrt(dt)

    # Poisson process (counting process with rate lam)
    # Inter-arrival exponential; here approximate with Poisson increments
    pois_increments = rng.poisson(lam*dt, size=n)
    pois_proc = np.cumsum(pois_increments)

    # Simple Markov chain with random transition matrix
    P = rng.random((Nstates, Nstates))
    P = P / P.sum(axis=1, keepdims=True)
    states = np.zeros(n, dtype=int)
    for k in range(n-1):
        states[k+1] = rng.choice(Nstates, p=P[states[k]])
    # Random walk
    steps = rng.standard_normal(n)
    walk = np.cumsum(steps) * np.sqrt(dt)

    return t, white, x, gauss_proc, pois_proc, states, walk

def plot_processes(t, white, x, gauss_proc, pois_proc, states, walk):
    fig, ax = plt.subplots()
    ax.plot(t, x, label='OU')
    ax.set_xlabel('t'); ax.set_ylabel('x(t)')
    ax.set_title('Ornstein–Uhlenbeck (OU)')
    ax.legend(frameon=False)
    plt.show()

    fig, ax = plt.subplots()
    ax.plot(t, white)
    ax.set_xlabel('t'); ax.set_ylabel('white')
    ax.set_title('White Noise')
    plt.show()

    fig, ax = plt.subplots()
    ax.plot(t, gauss_proc)
    ax.set_xlabel('t'); ax.set_ylabel('G(t)')
    ax.set_title('Gaussian Process (Integrated White Noise)')
    plt.show()

    fig, ax = plt.subplots()
    ax.plot(t, pois_proc)
    ax.set_xlabel('t'); ax.set_ylabel('N(t)')
    ax.set_title('Poisson Counting Process')
    plt.show()

    fig, ax = plt.subplots()
    ax.plot(t, states)
    ax.set_xlabel('t'); ax.set_ylabel('state')
    ax.set_title('Finite-State Markov Chain')
    plt.show()

    fig, ax = plt.subplots()
    ax.plot(t, walk)
    ax.set_xlabel('t'); ax.set_ylabel('W(t)')
    ax.set_title('Random Walk / Brownian Approximation')
    plt.show()

def run_demo(T=10.0, dt=0.01, theta=1.0, sigma=1.0, mu=0.0, lam=5.0, Nstates=3):
    t, white, x, gauss_proc, pois_proc, states, walk = simulate_processes(T, dt, theta, sigma, mu, lam, Nstates)
    plot_processes(t, white, x, gauss_proc, pois_proc, states, walk)

if HAS_WIDGETS:
    ui = W.VBox([
        W.FloatSlider(value=10.0, min=1.0, max=60.0, step=1.0, description='T'),
        W.FloatLogSlider(value=0.01, base=10, min=-3, max=-1, step=0.1, description='dt'),
        W.FloatSlider(value=1.0, min=0.0, max=5.0, step=0.1, description='theta'),
        W.FloatSlider(value=1.0, min=0.0, max=5.0, step=0.1, description='sigma'),
        W.FloatSlider(value=0.0, min=-2.0, max=2.0, step=0.1, description='mu'),
        W.FloatSlider(value=5.0, min=0.0, max=20.0, step=0.5, description='lambda'),
        W.IntSlider(value=3, min=2, max=8, step=1, description='Nstates')
    ])
    out = W.interactive_output(
        lambda T, dt, theta, sigma, mu, lambda_, Nstates: run_demo(T, dt, theta, sigma, mu, lambda_, Nstates),
        {'T': ui.children[0], 'dt': ui.children[1], 'theta': ui.children[2], 'sigma': ui.children[3], 'mu': ui.children[4], 'lambda_': ui.children[5], 'Nstates': ui.children[6]}
    )
    display(ui, out)
else:
    # Static example if widgets not available
    run_demo()
````

# 6.3. Probability Density Functions


We review definitions and compute empirical PDFs from signals:
- **Definition:** A PDF \(p(x)\) satisfies \(p(x)\ge0\) and \(\int p(x)\,dx=1\).
- **Estimation:** Use histograms with careful binning; compare to analytic PDFs (Gaussian, Exponential, Gamma).
- **Normalization:** Verify unit area and explore sensitivity to bin width.
## Definition and Description of a Probability Density Function (PDF)

### Fundamental Definition

A **probability density function (PDF)** is a mathematical function that describes the likelihood of a **continuous random variable** \( X \) taking on values within a given range.

Formally, the PDF \( p(x) \) is defined such that the **probability** of \( X \) lying between two values \( a \) and \( b \) is given by:

$$
P(a \le X \le b) = \int_{a}^{b} p(x)\,dx.
$$

The PDF itself does not give probability directly — rather, **the area under the curve** between two points represents the probability of \( X \) falling within that interval.

---

### Properties of a Valid PDF

For any continuous random variable \( X \), the corresponding PDF \( p(x) \) satisfies:

1. **Non-negativity:**
   $$
   p(x) \ge 0 \quad \forall x.
   $$

2. **Normalization (Total Probability = 1):**
   $$
   \int_{-\infty}^{+\infty} p(x)\,dx = 1.
   $$

3. **Units:**
   Since \( p(x)\,dx \) is dimensionless (a probability), \( p(x) \) has units of \( 1/[\text{units of } x] \).

---

### Relationship to the Cumulative Distribution Function (CDF)

The **cumulative distribution function (CDF)**, denoted by \( F(x) \), represents the probability that the random variable \( X \) is less than or equal to a value \( x \):

$$
F(x) = P(X \le x) = \int_{-\infty}^{x} p(\xi)\,d\xi.
$$

The PDF is obtained by differentiating the CDF:

$$
p(x) = \frac{dF(x)}{dx}.
$$

---

### Physical Interpretation in Turbulence

In turbulence, many quantities (e.g., velocity fluctuations, temperature fluctuations, scalar concentrations, or dissipation rates) exhibit random variability.  
Their instantaneous values can be treated as outcomes of a random variable \( X \), whose PDF \( p(x) \) captures the statistical structure of the fluctuations.

For instance:
- A **Gaussian PDF** often describes velocity fluctuations in weakly turbulent or central-limit regimes.
- A **Gamma or Lognormal PDF** can represent positive-definite quantities such as dissipation rate or scalar concentration.
- Non-Gaussian tails in \( p(x) \) indicate **intermittency** and **rare, intense events** characteristic of developed turbulence.

---

### Empirical Estimation of a PDF

From data, \( p(x) \) can be estimated using histograms or kernel density estimation.  
If the sample consists of \( N \) observations \( \{x_i\} \), then the **empirical PDF** in a bin of width \( \Delta x \) centered at \( x_j \) is:

$$
p(x_j) \approx \frac{1}{N\,\Delta x}\, \sum_{i=1}^{N} \mathbf{1}_{|x_i - x_j| < \frac{\Delta x}{2}},
$$

where \( \mathbf{1} \) is the indicator function (1 if the condition is true, 0 otherwise).

When integrated over all bins, the empirical PDF also satisfies normalization:

$$
\sum_j p(x_j)\,\Delta x \approx 1.
$$

---

### Summary

| Concept | Symbol | Expression / Property |
|:--|:--:|:--|
| $$\text{Definition   |    p(x)   |  }  (P(a \le X \le b) = \int_a^b p(x)\,dx ) |$$
| $$\text{Non-negativity | }( p(x) \ge 0 ) \text{  |   |}$$
| $$\text{Normalization | }( \int_{-\infty}^{\infty} p(x)\,dx = 1 ) |  $$|
| $$\text{Relation to CDF | }( F(x) = \int_{-\infty}^{x} p(\xi)\,d\xi ) |  $$|
| $$\text{Derivative relation | }( p(x) = \dfrac{dF(x)}{dx} ) |  $$|
| Physical meaning | – | Probability density of a turbulent variable |
| Typical shapes | Gaussian, Gamma, Lognormal, etc. | Depend on flow regime and variable type |

---

> ** Note:**  
> The PDF provides a complete statistical description of a continuous random variable.  
> In turbulence research, analyzing PDFs of velocity, vorticity, or dissipation helps reveal underlying dynamics, intermittency, and deviation from equilibrium statistics.


# Example NB 6.2 : Computing an Empirical Probability Density Function (PDF)

This code cell demonstrates how to **estimate a probability density function (PDF)** directly from a set of sampled or simulated data.  
In turbulence and random process analysis, such empirical PDFs are essential for quantifying how frequently certain fluctuation amplitudes occur.

---

## What the Code Does

1. **Generate a Non-Gaussian Dataset**

A synthetic dataset \( x \) is created by concatenating two Gaussian distributions:
- \( 4000 \) samples drawn from \( \mathcal{N}(0,1) \) — a standard normal distribution.
- \( 1000 \) samples drawn from \( \mathcal{N}(3,0.5^2) \) — a narrower distribution shifted to the right.

The resulting mixture exhibits **non-Gaussian behavior**, with a distinct secondary peak and asymmetry.  
This mimics how turbulent fluctuations can depart from purely Gaussian statistics due to intermittency or multiple physical mechanisms.

---

2. **Compute the Empirical PDF**

The helper function `empirical_pdf()` divides the range of  x  into a specified number of bins and counts the number of samples in each bin using `numpy.histogram`.  
The bin counts are then normalized such that the total area under the histogram equals one, satisfying the PDF normalization condition:

$$
\int_{-\infty}^{\infty} p(x)\,dx = 1.
$$

This normalization ensures that each bar height ( p(x_j) \ represents the **probability density** per unit interval.

---

3. **Extract the Histogram Parameters**

- **`centers`**: midpoints of the histogram bins.  
- **`hist`**: normalized bin heights representing \( p(x_j) \).  
- **`edges`**: boundaries of the bins.  
- **`area`**: numerical verification of the normalization  
  $$
  \text{area} = \sum_j p(x_j)\,\Delta x_j \approx 1.
  $$

If the computed area is close to 1, it confirms that the histogram correctly approximates a PDF.

---

4. **Plot the Empirical PDF**

The resulting histogram is plotted as a bar chart with:
- ( x )-axis: random variable ( x),
- ( y )-axis: estimated probability density ( p(x) ),
- title displaying the computed area for normalization verification.

Each bar represents the **probability density** within a small interval of \( x \).  
For a continuous random variable, the probability that ( X ) lies between two bounds ( [a,b] ) is approximated by the sum (or integral) of bar areas between ( a ) and ( b ).

---

## Pedagogical Significance

- Illustrates how **sample data can be transformed into a continuous PDF**, a key step in statistical turbulence analysis.  
- Highlights how **non-Gaussian features** (such as secondary peaks or asymmetry) can be identified visually.  
- Provides the computational foundation for later comparisons between **empirical PDFs** and **analytic distributions** (e.g., Gaussian, Gamma, or Lognormal).

---

> **Summary:**  
> This example connects the theoretical definition  
> $$
 P(a \le X \le b) = \int_a^b p(x)\,dx
> $$  
> with its **numerical estimation** from data using a histogram, forming the bridge between **mathematical probability** and **experimental or simulated turbulence statistics**.


````python
#Example NB 6.2
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(123)

def empirical_pdf(x, bins=50, density=True):
    hist, edges = np.histogram(x, bins=bins, density=density)
    centers = 0.5*(edges[:-1] + edges[1:])
    area = np.sum(hist*np.diff(edges))
    return centers, hist, edges, area

# Example signal: mixture to show non-Gaussianity
x = np.concatenate([rng.normal(0,1,4000), rng.normal(3,0.5,1000)])

centers, hist, edges, area = empirical_pdf(x, bins=60, density=True)

fig, ax = plt.subplots()
ax.bar(centers, hist, width=np.diff(edges), align='center')
ax.set_xlabel('x'); ax.set_ylabel('p(x)')
ax.set_title('Empirical PDF (Histogram) — Area ≈ {:.3f}'.format(area))
plt.show()
````

# 6.3 Statistical Moments of a Distribution


We compute the first four moments for given samples:
- **Mean** \(\mu = \mathbb{E}[X]\)
- **Variance** \(\sigma^2 = \mathbb{E}[(X-\mu)^2]\)
- **Skewness** \(\gamma_1 = \mathbb{E}[(X-\mu)^3]/\sigma^3\)
- **Kurtosis (excess)** \(\gamma_2 = \mathbb{E}[(X-\mu)^4]/\sigma^4 - 3\)


Statistical moments describe the **shape and characteristics** of a probability distribution.  
They provide compact measures of **location, spread, asymmetry,** and **tailedness**, all of which are essential for interpreting turbulent fluctuations.

For a random variable \( X \) with probability density function \( p(x) \), the \( n^\text{th} \) moment about the mean is defined as:

$$
m_n = \mathbb{E}\!\left[(X - \mu)^n\right] = \int_{-\infty}^{\infty} (x - \mu)^n\,p(x)\,dx
$$

where \( \mu = \mathbb{E}[X] \) is the mean value.

---

## The First Four Moments

### 1. Mean — Central Tendency

$$
\mu = \mathbb{E}[X]
$$

The mean represents the **expected or average value** of the random variable.  
In turbulence, the mean velocity \( \overline{U} \) defines the steady component about which fluctuations occur.

**Significance:**
- Quantifies the central or equilibrium state.  
- Separates deterministic (mean) and stochastic (fluctuating) components.  
- Used as a reference point for higher moments.

---

### 2. Variance — Measure of Dispersion

$$
\sigma^2 = \mathbb{E}\!\left[(X - \mu)^2\right]
$$

The variance quantifies the **spread or intensity of fluctuations** about the mean.  
In turbulence, the variance of velocity fluctuations corresponds to the **turbulent kinetic energy (TKE)** per unit mass.

**Significance:**
- Indicates the energy content of turbulent fluctuations.  
- A larger variance implies stronger deviations from the mean flow.  
- Often represented as \( u'^2, v'^2, w'^2 \) in experimental data.

$$

$$

The skewness describes the **degree of asymmetry** of the probability distribution about the mean.

- \( \gamma_1 < 0 \): the distribution has a longer left tail (rare large negative events).

- Reveals directional bias or intermittency in turbulent fluctuations.  
- In shear or buoyant turbulence, skewness indicates preferential transport (e.g., upward vs. downward events).  
- Non-zero skewness is a hallmark of **non-Gaussian** turbulence behavior.

---

### 4. Kurtosis (Excess) — Measure of Peakedness or Tail Heaviness

# Example NB6.4: Calculation of moments for Sample function


This cell **computes and prints** the first four **sample** statistics from an array ( x ) and, if ( x ) is not already defined, **creates** a non-Gaussian test signal.

**1) Data source (conditional):**
- If a variable `x` already exists (e.g., from the PDF example), it is reused.
- Otherwise, a synthetic mixture is generated:
  - ( 4000 ) samples from ( \mathcal{N}(0,1) ) and ( 1000 ) samples from \( \mathcal{N}(3, 0.5^2) \),
  creating a clearly **non-Gaussian** dataset.

**2) Computation performed by `sample_moments(x)`:**
- Converts input to a NumPy array and forms centered data ( c = x - \mu \).
- Computes **central moments** (with population denominators):
  $$
  \mu = \frac{1}{N}\sum_{i=1}^N x_i, \qquad
  m_2 = \frac{1}{N}\sum c_i^2, \qquad
  m_3 = \frac{1}{N}\sum c_i^3, \qquad
  m_4 = \frac{1}{N}\sum c_i^4
  $$
- Returns:
  - mean \( \mu \)
  - variance \( m_2 \)
  - skewness \( \displaystyle \frac{m_3}{m_2^{3/2}} \)
  - excess kurtosis \( \displaystyle \frac{m_4}{m_2^2} - 3 \)

**3) Numerical detail:**
- A small term \( 10^{-12} \) is added in denominators to avoid divide-by-zero when \( m_2 \to 0 \).

**4) Output:**
- Prints four lines:
  ```
  Mean     : <value>
  Variance : <value>
  Skewness : <value>
  Kurtosis : <value>   # excess kurtosis (Gaussian ≈ 0)
  ```

**Notes:**
- Variance here is the **population** estimate (dividing by \( N \), not \( N-1 \)).  
- Skewness and kurtosis are **standardized** central-moment forms; kurtosis reported is **excess**.
- The generated mixture ensures visible non-Gaussian values (nonzero skewness and excess kurtosis).


````python
import numpy as np

def sample_moments(x):
    x = np.asarray(x)
    mu = x.mean()
    c = x - mu
    m2 = np.mean(c**2)
    m3 = np.mean(c**3)
    m4 = np.mean(c**4)
    skew = m3 / (m2**1.5 + 1e-12)
    kurt_ex = m4 / (m2**2 + 1e-12) - 3.0
    return mu, m2, skew, kurt_ex

# Reuse x from previous cell if available; else generate
try:
    x
except NameError:
    rng = np.random.default_rng(123)
    x = np.concatenate([rng.normal(0,1,4000), rng.normal(3,0.5,1000)])

mu, var, skew, kurt = sample_moments(x)
print(f"Mean     : {mu:.4f}")
print(f"Variance : {var:.4f}")
print(f"Skewness : {skew:.4f}")
print(f"Kurtosis : {kurt:.4f}")
````

# 6.4  Stationarity and Ergodicity in Stochastic Processes

In the study of turbulence and random processes, it is crucial to distinguish between two related but distinct concepts: **stationarity** and **ergodicity**.  
Both describe how the statistical properties of a random signal behave in time and across ensembles.

---

## 1. Ensemble Mean — Definition and Physical Interpretation

In probability theory and turbulence statistics, the **ensemble mean** is the *expected value* of a stochastic process at a given instant in time, computed over many possible realizations (or experiments).


$$\text{Consider a random process } ( X(t, \omega) ), \text{  
where t  is time and }  \omega \text{labels one realization (or outcome) within the ensemble}$$.

The **ensemble mean** is defined as

$$
\langle X(t) \rangle_{\text{ensemble}}
\;=\;
\mathbb{E}[X(t)]
\;=\;
\int_{-\infty}^{\infty} x\,p(x, t)\,dx
\;=\;
\lim_{M \to \infty} \frac{1}{M}\sum_{m=1}^{M} X_m(t),
$$

where:
- \( X_m(t) \) is the value of the process at time \( t \) for the \( m^{\text{th}} \) realization,
- \( M \) is the total number of realizations in the ensemble,
- \( p(x, t) \) is the probability density function of \( X \) at time \( t \).

---

- The **ensemble mean** is what you would get if you could repeat the same experiment many times under *identical conditions* and then average all those outcomes at the same time instant.
- It represents the “expected value” or “mean behavior” of the process across all possible worlds, not just one observation.

For example:
- In a wind-tunnel measurement, if you could restart the flow 1000 times with the same boundary conditions and record the instantaneous velocity at time \( t_0 \),  
  the average of those 1000 velocity readings would approximate the **ensemble mean** \( \langle U(t_0) \rangle \).

---

### 3. Connection to Time Averaging and Ergodicity

The **time average** over one long realization is

$$
\overline{X}_{\text{time}}
= \frac{1}{T} \int_0^T X(t)\,dt.
$$

If the process is **ergodic**, then for large \( T \):

$$
\overline{X}_{\text{time}}
\;\approx\;
\langle X(t) \rangle_{\text{ensemble}}.
$$

Thus, ergodicity allows us to replace the *ensemble mean* (an average over many experiments)  
with a *time average* from a single, sufficiently long record.

---

In turbulence statistics:
- The **ensemble mean velocity** \( \langle U_i \rangle \) represents the expected flow field over many possible realizations of the same flow configuration.
- Because turbulence is usually assumed to be **stationary and ergodic**, we approximate the ensemble mean using a **time average**:

$$
\langle U_i \rangle \;\approx\; \overline{U_i}
= \frac{1}{T} \int_0^T U_i(t)\,dt.
$$

This is the foundation of the **Reynolds decomposition**:

$$
U_i(t) = \overline{U_i} + u_i'(t),
$$

where \( \overline{U_i} \) is the (time or ensemble) mean and \( u_i'(t) \) is the fluctuation.

---


## 2. Stationarity

A stochastic process ( X(t)) is said to be **stationary** if its statistical properties do not change with time.

Formally, for any time shift ( \tau \),
$$
\mathbb{E}[X(t)] = \mathbb{E}[X(t+\tau)], \qquad
\mathrm{Var}[X(t)] = \mathrm{Var}[X(t+\tau)],
$$
and more generally,
$$
\text{Cov}[X(t_1), X(t_2)] = R_X(t_2 - t_1),
$$
i.e., the covariance depends only on the time difference, not the absolute time.

**Interpretation:**  
A stationary process “looks statistically the same” no matter when you observe it.  
In turbulence, a stationary time series might represent velocity fluctuations measured in a statistically steady flow.

---

## 3. Ergodicity

A process is **ergodic** if the **time averages** along one sufficiently long realization are equal to the **ensemble averages** over many realizations:
$$
\overline{X}_{\text{time}} = \lim_{T\to\infty} \frac{1}{T}\int_0^T X(t)\,dt
\quad \text{equals} \quad
\langle X \rangle_{\text{ensemble}} = \mathbb{E}[X].
$$

This means that a **single, long record** can represent the behavior of the whole ensemble.  
Ergodicity is what allows experimentalists and numerical modelers to compute statistics (mean, variance, spectra) from one measurement rather than an infinite set.

---

### 3. Relationship Between Stationarity and Ergodicity

- **Stationary but not ergodic:**  
  The process has constant statistical properties, but one realization may not represent the ensemble (e.g., two stationary states with different means).

- **Ergodic implies stationarity (usually with extra conditions):**  
  If time averages always converge to ensemble averages, the process must have stationary statistics.

---

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(t, one_realization, alpha=0.5, label='One realization')
ax.plot(t, ensemble_mean, 'k--', linewidth=2, label='Ensemble mean')
ax.axhline(time_avg, color='r', linestyle=':', linewidth=2, label='Time average (1 realization)')
ax.set_xlabel('t'); ax.set_ylabel('X(t)')
ax.set_title('Ergodic Behavior: Time Average ≈ Ensemble Average')
ax.legend(frameon=False)
plt.show()
```

**Observation:**
- For Gaussian white noise (stationary), the time average of one realization approximates the ensemble average — confirming ergodicity.
- In contrast, if the mean were drifting (non-stationary), the time average and ensemble average would differ.

---

### 6. Summary Table

| Concept | Definition | Example | Implication |
|:--|:--|:--|:--|
| **Stationarity** | Statistics invariant with time | White noise with fixed variance | Required for ergodicity |
| **Ergodicity** | Time averages equal ensemble averages | Long velocity record in steady turbulence | Allows single-record statistics |
| **Non-Stationary** | Mean or variance changes in time | Signal with drift or periodic forcing | Ensemble ≠ time statistics |

---

### 7. Physical Relevance in Turbulence

In experimental or simulated turbulence:
- A **stationary** flow means the mean and fluctuation statistics (TKE, Reynolds stress) have settled to steady values.  
- **Ergodicity** allows computing ensemble-averaged quantities (e.g., mean velocity, RMS fluctuations) from a single long signal rather than repeated experiments.

> **Key insight:**  
> In turbulence research, we assume that sufficiently long, statistically steady records are *ergodic* and *stationary* — enabling us to treat temporal averages as ensemble averages in the Reynolds decomposition framework.


````python
# --- Ergodicity: interactive demo (white vs OU) ---
import numpy as np
import matplotlib.pyplot as plt

try:
    import ipywidgets as W
    from IPython.display import display
    HAS_WIDGETS = True
except Exception:
    HAS_WIDGETS = False

_rng = np.random.default_rng(314)

def _simulate_white(N, sigma):
    return _rng.normal(0.0, sigma, N)

def _simulate_ou(N, dt, theta, sigma, mu=0.0):
    x = np.zeros(N); x[0] = mu
    for k in range(N-1):
        x[k+1] = x[k] + (-theta*(x[k]-mu))*dt + sigma*np.sqrt(dt)*_rng.standard_normal()
    return x

def _ergodicity_demo(proc='White', N=2000, M=50, sigma=1.0, theta=1.0, dt=0.01):
    # Build ensemble
    if proc == 'White':
        ensemble = np.stack([_simulate_white(N, sigma) for _ in range(M)], axis=0)
    else:
        ensemble = np.stack([_simulate_ou(N, dt, theta, sigma) for _ in range(M)], axis=0)

    # Ensemble mean at each time
    ens_mean = ensemble.mean(axis=0)

    # One realization + its time avg
    one = ensemble[0]
    time_avg = one.mean()

    t = np.arange(N) * dt
    fig, ax = plt.subplots(figsize=(9,4))
    ax.plot(t, one, alpha=0.5, label='one realization')
    ax.plot(t, ens_mean, 'k--', linewidth=2, label='ensemble mean')
    ax.axhline(time_avg, color='r', linestyle=':', linewidth=2, label='time average (one)')
    ax.set_xlabel('t'); ax.set_ylabel('X(t)')
    ax.set_title(f'Ergodicity demo — {proc} process')
    ax.legend(frameon=False)
    plt.show()

if HAS_WIDGETS:
    ui = W.VBox([
        W.Dropdown(options=['White','OU'], value='White', description='Process'),
        W.IntSlider(value=2000, min=200, max=20000, step=200, description='N (samples)'),
        W.IntSlider(value=50, min=5, max=200, step=5, description='M (realizations)'),
        W.FloatSlider(value=1.0, min=0.1, max=3.0, step=0.05, description='σ (noise)'),
        W.FloatSlider(value=1.0, min=0.1, max=5.0, step=0.1, description='θ (OU only)'),
        W.FloatLogSlider(value=0.01, base=10, min=-3, max=-1, step=0.1, description='Δt')
    ])
    def _update(proc, N, M, σ, θ, Δt):
        _ergodicity_demo(proc, N, M, σ, θ, Δt)
    out = W.interactive_output(_update, {
        'proc': ui.children[0], 'N': ui.children[1], 'M': ui.children[2],
        'σ': ui.children[3], 'θ': ui.children[4], 'Δt': ui.children[5]
    })
    display(ui, out)
else:
    print("ipywidgets not available — showing a static OU example. Install ipywidgets to enable interactivity.")
    _ergodicity_demo(proc='OU', N=3000, M=80, sigma=1.0, theta=1.0, dt=0.01)
````

````python
### Interpreting the Interactive Plots

These interactive demonstrations visualize how **stationarity** and **ergodicity** appear in real data or synthetic random signals.

---

#### 1. Stationarity vs. Non-Stationarity

**Top Left — Stationary Process (White Noise):**
- The signal fluctuates randomly around zero.
- Its amplitude envelope and fluctuation level remain statistically constant over time.
- This means that if you compute its mean and variance over any time segment, the values will be roughly the same.
- In turbulence, this corresponds to a *statistically steady* flow — e.g., velocity fluctuations in a fully developed channel flow.

**Top Right — Non-Stationary Process (With Drift):**
- The signal includes a slow upward drift (a deterministic trend).
- The random component is still present, but the mean value is changing with time.
- This violates the condition of stationarity because the expected value depends on time, \( \mathbb{E}[X(t)] \neq \mathbb{E}[X(t+\tau)] \).

**Bottom Row — Rolling Mean and Variance:**
- The rolling mean (blue) and rolling variance (orange) are computed over a moving time window.
- For the stationary process, both remain nearly constant (small fluctuations only).
- For the non-stationary process, the rolling mean increases steadily, and sometimes the variance also changes, showing **time-evolving statistics**.

**Interpretation:**
> - Constant rolling statistics → **stationary**  
> - Changing rolling statistics → **non-stationary**  
> - Stationarity implies that statistical properties (mean, variance, higher moments) depend only on time difference, not absolute time.

**Turbulence Analogy:**
In a wind-tunnel experiment, if you record velocity in a stable operating condition, the fluctuations will look stationary.  
If the fan speed slowly increases during the test, the signal becomes non-stationary.

---

#### 2. Ergodicity Demonstration

**What’s Shown:**
- The black dashed curve is the **ensemble mean**, i.e., the average across many independent realizations of the same stochastic process.  
- The light blue line is a **single realization** (one experiment or simulation).  
- The red dotted line is the **time average** of that single realization.

**Interpretation:**
- If the process is **ergodic**, then for large sample sizes:
  - The **time average** of one realization ≈ the **ensemble average** across many realizations.
  - That is, \( \overline{X}_{\text{time}} \approx \langle X \rangle_{\text{ensemble}} \).
- For the white-noise or OU (Ornstein–Uhlenbeck) process, both are stationary and ergodic:
  - The red line (time average) and black dashed line (ensemble mean) overlap closely.  
  - This shows that one long time series contains all the statistical information needed.
- If you were to make the process non-stationary (e.g., add drift), the red line would diverge from the ensemble mean, showing **loss of ergodicity**.

**Physical Meaning in Turbulence:**
- In a **stationary turbulent flow**, one long time record of velocity \( u'(t) \) can represent the entire ensemble of possible outcomes — this is the assumption behind time-averaged Reynolds equations.
- In **transient or evolving turbulence**, ergodicity breaks down: each record represents a different state, so ensemble averaging must be done explicitly (e.g., over multiple experiments).

---

#### 3. Summary

| Concept | What the Plot Shows | Implication |
|:--|:--|:--|
| **Stationary process** | Rolling mean and variance nearly constant | Statistics independent of time; steady turbulence |
| **Non-stationary process** | Mean drifts or variance changes | Statistics evolve; not suitable for Reynolds averaging |
| **Ergodic process** | Time and ensemble averages coincide | Single long record = full statistical representation |
| **Non-ergodic process** | Time and ensemble averages differ | Need multiple realizations to compute averages |

---

**Key Takeaway:**  
> In turbulence and stochastic systems, we rely on **stationarity** and **ergodicity** to justify computing averages from limited data.  
> The plots make these abstract ideas visible — showing that *steady randomness* (stationarity) and *representative time histories* (ergodicity) are the foundations of statistical turbulence analysis.
````
